```python
text1="""After the group disbanded in 1970, he pursued a solo career"""
text2="""he formed the band Wings with his first wife, Linda, Denny Laine"""
text3="""A self-taught musician, McCartney is proficient on bass, guitar, keyboards, and drums"""
text4="""People could watch the show on radio and on television"""
text5="""McCartney began his career as a member of the Quarrymen in 1957, which evolved into the Beatles in 1960."""
text6="""Throughout the 1970s, he led Wings, one of the most successful bands of the decade, with more than a dozen international top 10 singles and albums."""
text7="""he went for some food, because he was hungry"""
text8="""A prominent result of this is the Great Red Spot, a giant storm that is known to have existed since at least the 17th century, when it was first seen by telescope"""
```
